```python
# Q1: Date cleanup
from dateutil import parser

# Load dataset
df = pd.read_csv(r'C:\Users\Akshiya George\OneDrive\Desktop\Data Science\Amazon\amazon_india_2015.csv')

# cleaning function
def clean_date(date_input):
    try:
        # Convert to string safely without using str() directly
        date_str = "{}".format(date_input).strip()
        parsed_date = parser.parse(date_str, dayfirst=True)
        return parsed_date.strftime('%Y-%m-%d')
    except Exception:
        return pd.NaT

# Apply function to the column
df['order_date'] = df['order_date'].apply(clean_date)

# Save the cleaned data back to the same file
df.to_csv(r'C:\Users\Akshiya George\OneDrive\Desktop\Data Science\Amazon\amazon_india_2015.csv', index=False)

print("✅ Dates cleaned and saved successfully.")
```

```python
# Q2:Clean price column
import re

# Load dataset
df = pd.read_csv(r'C:\Users\Akshiya George\OneDrive\Desktop\Data Science\Amazon\amazon_india_2015.csv')

# Define the cleaning function
def clean_price(value):
    try:
        value = str(value).strip()
        # Remove currency symbols and non-numeric characters
        value = re.sub(r'[₹Rsâ‚¹]', '', value, flags=re.IGNORECASE)
        value = value.replace(',', '')
        # Remove non-numeric entries
        if not re.match(r'^-?\d+(\.\d+)?$', value):
            return pd.NA
        return abs(float(value))  # Convert negative to positive
    except:
        return pd.NA

# Apply the function to the column
df['original_price_inr'] = df['original_price_inr'].apply(clean_price)

# Save the cleaned data back to the same file
df.to_csv(r'C:\Users\Akshiya George\OneDrive\Desktop\Data Science\Amazon\amazon_india_2015.csv', index=False)

print("✅ Price column cleaned, negatives converted to positives, and file saved successfully.")
```

```python
# Q3: Standardize ratings
# Load your dataset
df = pd.read_csv(r'C:\Users\Akshiya George\OneDrive\Desktop\Data Science\Amazon\amazon_india_2015.csv')

# Define the rating cleaner
def clean_rating(value):
    try:
        value = str(value).strip().lower()

        if value in ['', 'nan', 'none', 'null']:
            return 'NA'

        if 'star' in value:
            match = re.search(r'(\d+(\.\d+)?)', value)
            return float(match.group(1)) if match else 'NA'

        if '/' in value:
            parts = value.split('/')
            if len(parts) == 2 and float(parts[1]) != 0:
                return round(float(parts[0]) / float(parts[1]) * 5, 1)
            else:
                return 'NA'

        if re.match(r'^\d+(\.\d+)?$', value):
            return float(value)

        return 'NA'
    except:
        return 'NA'

# Apply the function to the column
df['customer_rating'] = df['customer_rating'].apply(clean_rating)

# Save the updated file
df.to_csv(r'C:\Users\Akshiya George\OneDrive\Desktop\Data Science\Amazon\amazon_india_2015.csv', index=False)

print("✅ Missing ratings filled with 'NA' and file updated successfully.")
```
